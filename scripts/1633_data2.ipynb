```python
%matplotlib notebook
plt.rcParams['figure.figsize']=(6,5)
plt.rcParams['figure.dpi']=100

fig=plt.figure()

x='AP_HW'
y='AP_500'
z='f_AHP'

ax= fig.add_subplot (111, projection= '3d')
ax.scatter (xs=data2[x], ys=data2[y], zs=data2[z],
           s=80, c='black', marker='o', edgecolors= 'black', depthshade= False, label='FS_cells')
ax.scatter (xs=data4[x], ys=data4[y], zs=data4[z],
           s=80, c='white', marker='o', edgecolors= 'black', depthshade= False, label='Pyramidal_cells')

ax.set_xlabel(x)
ax.set_ylabel(y)
ax.set_zlabel(z)

ax.legend()
plt.show()
fig.savefig("3d_figure_6.pdf", bbox_inches='tight')
```
