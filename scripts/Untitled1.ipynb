```python
%matplotlib notebook
plt.rcParams['figure.figsize']=(5,4)
plt.rcParams['figure.dpi']=150

fig=plt.figure()

x='AP_HW'
y='AP_500'
z='f_AHP'

ax= fig.add_subplot (111, projection= '3d')
ax.scatter (xs=old[x], ys=old[y], zs=old[z],
           s=80, c='black', marker='o', edgecolors= 'black', depthshade= True, label='FS_cells')
ax.scatter (xs=new[x], ys=new[y], zs=new[z],
           s=80, c='white', marker='o', edgecolors= 'black', depthshade= True, label='Pyramidal_cells')

ax.set_xlabel(x)
ax.set_ylabel(y)
ax.set_zlabel(z)

ax.legend()
plt.show()
```
